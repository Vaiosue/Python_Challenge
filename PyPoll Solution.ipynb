```python
#Import dependencies

import os
import csv
```

```python
# set the total number of votes before the election begins equal to zero

tot_votes = 0

# list the names of the candidates up for election

cand_list = []

cand_name = []

# set the vote counter and vote percentage for 4 candidates to zero

cand_vote = [0, 0, 0, 0]
cand_vote_percent = [0, 0, 0, 0]
cand_winner = []
```

```python
# Set the file path to the election_data.csv

csvpath = os.path.join('..', 'Resources', 'election_data.csv')

# Open the csv

with open(csvpath, 'r', newline='') as csvfile:

    # Use csv.reader to open csv file and set to variable name csv_elec

    csv_elec = csv.reader(csvfile, delimiter=',')

    # define the header of csv_elec

    csv_header = next(csv_elec)
```

```python
# loop through entire file to count total votes, add names to the candidate_name list, and votes per candidate

    for row in csv_elec:
        total_votes += 1
        cand_list.append(str(row[2]))
    for row[2] in cand_list:
        if row[2] not in cand_name:
            cand_name.append(row[2])
        if row[2] == cand_name[0]:
            cand_vote[0] += 1
        elif row[2] == cand_name[1]:
            cand_vote[1] += 1
        elif row[2] == cand_name[2]:
            cand_vote[2] += 1
        elif row[2] == cand_name[3]:
            cand_vote[3] += 1

# Calculate each candidates percentage of the total vote

    cand_vote_percent[0] = round(100 * (cand_vote[0] / tot_votes), 4)
    cand_vote_percent[1] = round(100 * (cand_vote[1] / tot_votes), 4)
    cand_vote_percent[2] = round(100 * (cand_vote[2] / tot_votes), 4)
    cand_vote_percent[3] = round(100 * (cand_vote[3] / tot_votes), 4)

    # Calculate winner
    if cand_vote[0] == max(cand_vote[0], cand_vote[1], cand_vote[2], cand_vote[3]):
       cand_winner = cand_name[0]
    elif cand_vote[1] == max(cand_vote[0], cand_vote[1], cand_vote[2], cand_vote[3]):
       cand_winner = cand_name[1]
    elif cand_vote[2] == max(cand_vote[0], cand_vote[1], cand_vote[2], cand_vote[3]):
       cand_winner = cand_name[2]
    elif cand_vote[3] == max(cand_vote[0], cand_vote[1], cand_vote[2], cand_vote[3]):
       cand_winner = cand_name[3]

    # Print

    print("Election Results")
    print("-----------------------------")
    print(f"Total Votes: {tot_votes}")
    print("-----------------------------")
    print(f"{cand_name[0]}: {cand_vote_percent[0]}% ({cand_vote[0]})")
    print(f"{cand_name[1]}: {cand_vote_percent[1]}% ({cand_vote[1]})")
    print(f"{cand_name[2]}: {cand_vote_percent[2]}% ({cand_vote[2]})")
    print(f"{cand_name[3]}: {cand_vote_percent[3]}% ({cand_vote[3]})")
    print("-----------------------------")
    print(f"Winner: {cand_winner}")

# Set path for output file

output_path = os.path.join("..", "Output", "Election_Results.txt")
with open(output_path, 'w', newline='') as text_file:

    # Create text_file report output

    print("Election Results", file=text_file)
    print("-----------------------------", file=text_file)
    print(f"Total Votes: {tot_votes}", file=text_file)
    print("-----------------------------", file=text_file)
    print(f"{cand_name[0]}: {cand_vote_percent[0]}% ({cand_vote[0]})", file=text_file)
    print(f"{cand_name[1]}: {cand_vote_percent[1]}% ({cand_vote[1]})", file=text_file)
    print(f"{cand_name[2]}: {cand_vote_percent[2]}% ({cand_vote[2]})", file=text_file)
    print(f"{cand_name[3]}: {cand_vote_percent[3]}% ({cand_vote[3]})", file=text_file)
    print("-----------------------------", file=text_file)
    print(f"Winner: {cand_winner}", file=text_file)
```
